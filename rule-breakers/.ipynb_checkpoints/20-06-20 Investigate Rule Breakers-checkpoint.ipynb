# Data Cleaned

```python
fool = pd.read_csv('20-06-20_cleaned_rule_breakers.csv', index_col=0)
fool['REC DATE'] = pd.to_datetime(fool['REC DATE'])
fool
```

# Analyze Rule Breakers

```python
# investing $1000 each time for each of the 211 recomendation.
print(f"Rule Breakers: ${1000 * (len(fool['RETURN']) + fool['RETURN'].sum()):.0f}")
```

An intial \\$211,000 turns into \\$994,849

```python
rb_portfolio_return = {}

for ticker, return_amt in fool[['TICKER', 'RETURN']].values:
    rb_portfolio_return[ticker] = 1000 * (1 + return_amt) + rb_portfolio_return.get(ticker, 0)

rb_portfolio_return = {k: v for k, v in sorted(rb_portfolio_return.items(), key=lambda item: item[1], reverse=True)}
rb_portfolio_return
```

```python
labels = list(rb_portfolio_return.keys())[:5] + ['' for _ in list(rb_portfolio_return.keys())[5:]]
sizes = rb_portfolio_return.values()
fig1, ax1 = plt.subplots()
ax1.pie(sizes, labels=labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

ax1.set_title('Rule Breakers\'s Portfolio Proportion', fontsize=15);

plt.show()
```

# Analyze David, Normalized to Rule Breakers

```python
stock_advisor = pd.read_csv('../stock_advisor/20-06-20_cleaned_fool.csv', index_col=0)
stock_advisor['REC DATE'] = pd.to_datetime(stock_advisor['REC DATE'])
stock_advisor
```

```python
david = stock_advisor[stock_advisor['TEAM'] == 'David']
david_matched_to_rule_breakers = david[david['REC DATE'] >= datetime.datetime.strptime("12/15/04", "%m/%d/%y")]
david_matched_to_rule_breakers
```

```python
# investing $1750 each time for each of the 120 recomendation.
print(f"David: ${1750 * (len(david_matched_to_rule_breakers['RETURN']) + david_matched_to_rule_breakers['RETURN'].sum()):.0f}")
```

An intial \\$210,000 turns into \\$1,621,559. This is way better.

```python
david_portfolio_return = {}

for ticker, return_amt in david_matched_to_rule_breakers[['TICKER', 'RETURN']].values:
    david_portfolio_return[ticker] = 1000 * (1 + return_amt) + david_portfolio_return.get(ticker, 0)

david_portfolio_return = {k: v for k, v in sorted(david_portfolio_return.items(), key=lambda item: item[1], reverse=True)}
david_portfolio_return
```

# Returns and Normailzed Returns?

```python
fool.sort_values(by='RETURN', ascending=False).head(20)
```

```python
my_fool = fool.copy()
my_fool['Days to 06/26/20'] = fool['REC DATE'].apply(lambda x: (datetime.datetime.strptime("06/26/20", "%m/%d/%y") - x).days)
my_fool['Return over Days'] = my_fool['RETURN'] / my_fool['Days to 06/26/20']
my_fool
```

```python
my_fool.sort_values(by='Return over Days', ascending=False).head(20)
```
